# Set up file mapping

The working directory contains a pre-populated file called `dirs.txt` that contains all the directories on the merri cluster in which a feature map was created. (As well as the RGB images.) We use this to copy over each feature matrix.

````python
import subprocess as sp
import time

with open('dirs.txt', 'r') as fin:
    line = fin.readline().rstrip()
    dirs = line.split(' ')
    
def dir2plate(dirname):
    """Return a Plate ID from a directory name.
    
    Parameters
    ----------
    dirname : string
        A directory containing export images from an HCS plate.

    Returns
    -------
    plateid : string
        The plate ID parsed from the directory name.
    """
    basedir = os.path.split(dirname)[1]
    plateid = basedir.split('_')[1]
    return plateid

remote_base_dir = 'marcelle/raw-data'

plateids = map(dir2plate, dirs)
feature_fns = ['features-%s.h5' % p for p in plateids]
#for d, f in zip(dirs, feature_fns):
#    if not os.path.exists(f):
#        time.sleep(10) # merri appears to choke with too many scps
#        sp.check_call(['scp', 'jni@merri.vlsci.unimelb.edu.au:' +
#                 os.path.join(remote_base_dir, d, 'features.h5'), f])
````

# Initial exploratory analysis

Now, let's use scikit-learn to do some PCA.

````python
from sklearn.preprocessing import StandardScaler
from sklearn import decomposition
````

````python
X = StandardScaler().fit_transform(data)
pca = decomposition.PCA().fit(X)
pca.n_components = 2
X_reduced = pca.fit_transform(X)
print(X_reduced.shape)
````

The obligatory PCA scatter:

````python
plt.scatter(*(X_reduced.T))
````

What features contributed most to the two top PCA components?

````python
pca_comp1, pca_comp2 = (np.argsort(pca.components_[0])[-1:-4:-1],
                        np.argsort(pca.components_[1])[-1:-4:-1])
data.columns[pca_comp1], data.columns[pca_comp2]
````

# Clustering

Let's cluster the data (using KMeans):

````python
from sklearn import cluster
````

````python
# average cluster size of 10
avg_cluster_size = 20
K = cluster.MiniBatchKMeans(n_clusters=(X.shape[0] // avg_cluster_size))
K = K.fit(X)
````

The following function will return gene information from the .csv we loaded when given a cluster ID from the Ward clustering.

````python
def get_cluster_gene_data(cluster_id):
    plate_cluster = [(int(file_data[i]['plate']), file_data[i]['well']) for
                     i in np.flatnonzero(K.labels_ == cluster_id)]
    genes_cluster = [line for line in gene_data
                     if (int(line[2]), line[3]) in plate_cluster]
    return genes_cluster
````

# Grabbing the files from merri

````python
import mahotas as mh
def grab_image(plate, well):
    f = well2file[(plate, well)]
    f = (f.replace('w1_stitched', 'w?_stitched').
           replace('w2_stitched', 'w?_stitched').
           replace('w3_stitched', 'w?_stitched'))
    sp.check_call(['scp', 'jni@merri.vlsci.unimelb.edu.au:' + f,
                   '.'])
    time.sleep(1)
    try:
        f = f.replace('w?_stitched', 'w1_stitched')
        im = mh.imread(os.path.split(f)[1])
    except ValueError:
        try:
            f = f.replace('w1_stitched', 'w2_stitched')
            im = mh.imread(os.path.split(f)[1])
        except ValueError:
            f = f.replace('w2_stitched', 'w3_stitched')
            im = mh.imread(os.path.split(f)[1])
    return im
````

Now we want to look at the images. The below function shows three images side by side, but can be generalised to show a grid of any number of images.

````python
def show_3ims(ims):
    fig, axes = plt.subplots(1, 3, figsize=(15, 12))
    for im, ax in zip(ims, axes):
        ax.imshow(im)
    
````

````python
keys = [tuple(a[2:4]) for a in
        gene_cluster_1[-3:]]
try:
    ims = [grab_image(p, w) for
           p, w in keys]
except:
    print([well2file[k] for k in keys])
    raise
````

````python
show_3ims(ims)
````
